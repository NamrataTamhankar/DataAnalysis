```python
import pandas as pd

# Read the Parquet files into DataFrames
a = pd.read_parquet("C:/Users/a839429/Uber Data Analysis/Data/yellow_tripdata_2024-01.parquet")
b = pd.read_parquet("C:/Users/a839429/Uber Data Analysis/Data/yellow_tripdata_2024-02.parquet")

# Concatenate the DataFrames
merged = pd.concat([a, b])

# Write the merged DataFrame to a new Parquet file
parquet_file_path = "C:/Users/a839429/Uber Data Analysis/Data/2024.parquet"
merged.to_parquet(parquet_file_path, index=False)
```
